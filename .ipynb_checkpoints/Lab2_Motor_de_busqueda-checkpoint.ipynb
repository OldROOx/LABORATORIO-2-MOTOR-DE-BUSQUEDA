## 0 · Cargar el corpus procesado del Lab 1

Debe estar `corpus_procesado.json` en la misma carpeta. Si no, vuelvan a correr el Lab 1.

```python
import json, math

with open('corpus_procesado.json', encoding='utf-8') as fh:
    corpus = json.load(fh)

documentos = [d['tokens'] for d in corpus]          # lista de listas de terminos
print(f'{len(corpus)} documentos.  Ejemplo {corpus[0]["id"]}:', documentos[0][:8])
```

**Reutilicen su `preprocesar`.** Péguenla aquí *idéntica* a la del Lab 1. Es indispensable que la consulta pase por **exactamente el mismo pipeline** que el corpus (este es el error más común: vectorizar la consulta distinto y que nada coincida).

```python
import re, unicodedata
from collections import Counter

import nltk
nltk.download('punkt', quiet=True)
from nltk.stem import SnowballStemmer

import spacy
try:
    nlp = spacy.load('es_core_news_sm')
except OSError:
    from spacy.cli import download; download('es_core_news_sm')
    nlp = spacy.load('es_core_news_sm')

stopwords_es = nlp.Defaults.stop_words
CONSERVAR = {'no', 'mas', 'sin', 'nunca', 'jamás'}
MIS_STOPWORDS = set(stopwords_es) - CONSERVAR

def normalizar(texto, quitar_acentos=False):
    """Devuelve el texto normalizado (aun como string, sin tokenizar)."""
    texto = texto.lower()
    texto = unicodedata.normalize('NFC', texto)
    texto = re.sub(r'<[^>]+>', ' ', texto)
    texto = re.sub(r'https?://\S+', ' ', texto)
    texto = re.sub(r'[^\x00-\x7Fáéíóúüñ\s]', ' ', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    if quitar_acentos:
        texto = unicodedata.normalize('NFD', texto)
        texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn')
    return texto

def preprocesar(texto):
    """Pipeline definitivo: texto crudo -> lista de terminos limpios.
    Decisiones (identicas al Lab 1): quitar_acentos=True, MIS_STOPWORDS, lemmatizacion.
    """
    texto_norm = normalizar(texto, quitar_acentos=True)
    doc = nlp(texto_norm)
    tokens = [
        token.lemma_.lower()
        for token in doc
        if not token.is_punct
        and not token.is_space
        and not token.like_url
        and token.lemma_.lower() not in MIS_STOPWORDS
        and len(token.lemma_) > 1
        and token.lemma_.isalpha()
    ]
    return tokens
```

## 1 · Indexar — TF, IDF y TF-IDF desde cero

Implementen las tres funciones siguiendo las fórmulas de la clase. La estructura es la del slide de "15 líneas".
- `tf(doc)` → frecuencia **relativa**: `f(t,d) / |d|` (importancia local).
- `idf(corpus)` → `ln(N / df(t))` (rareza global). Cuenten **documentos**, no apariciones.
- `tfidf(doc, idf_)` → `tf(t,d) · idf(t)`.


```python
def tf(doc):
    """doc: lista de terminos -> dict termino:tf (frecuencia relativa)."""
    n = len(doc)
    if n == 0:
        return {}
    conteo = Counter(doc)
    return {termino: c / n for termino, c in conteo.items()}

def idf(corpus):
    """corpus: lista de docs (cada uno lista de terminos) -> dict termino:idf."""
    N = len(corpus)
    df = {}
    for doc in corpus:
        for termino in set(doc):          # set: contamos documentos, no apariciones
            df[termino] = df.get(termino, 0) + 1
    return {termino: math.log(N / df_t) for termino, df_t in df.items()}

def tfidf(doc, idf_):
    """-> dict termino:peso tf-idf para un documento."""
    tf_doc = tf(doc)
    return {termino: peso * idf_.get(termino, 0.0) for termino, peso in tf_doc.items()}
```

```python
# Construir el indice: un vector tf-idf (dict) por documento
IDF = idf(documentos)
INDICE = [tfidf(doc, IDF) for doc in documentos]

# Sanity check: el termino mas pesado de d04 (sequia/maiz) deberia ser tematico
import operator
top = sorted(INDICE[3].items(), key=operator.itemgetter(1), reverse=True)[:5]
print('Terminos top de', corpus[3]['id'], '->', top)
```

## 2 · Procesar la consulta

La consulta es texto libre. Pásenla por el **mismo** `preprocesar`, conviértanla en vector con el **mismo** `IDF` del corpus (no recalculen IDF con la consulta dentro).

```python
def vectorizar_consulta(texto):
    """texto libre -> vector tf-idf (dict) usando el IDF del corpus."""
    tokens = preprocesar(texto)
    tf_consulta = tf(tokens)
    return {termino: peso * IDF.get(termino, 0.0) for termino, peso in tf_consulta.items()}

print(vectorizar_consulta('sequia en los cultivos'))
```

## 3 · Ranquear — similitud coseno

Implementen el coseno entre dos vectores dispersos (dicts) y la función `buscar` que devuelve el **top-k**.
$$\cos(\vec{q},\vec{d}) = \frac{\vec{q}\cdot\vec{d}}{\lVert\vec{q}\rVert\,\lVert\vec{d}\rVert}$$

```python
def coseno(v1, v2):
    """Similitud coseno entre dos vectores dispersos (dicts). Maneja el caso de norma 0."""
    claves_comunes = set(v1.keys()) & set(v2.keys())
    producto_punto = sum(v1[t] * v2[t] for t in claves_comunes)
    norma1 = math.sqrt(sum(p ** 2 for p in v1.values()))
    norma2 = math.sqrt(sum(p ** 2 for p in v2.values()))
    if norma1 == 0 or norma2 == 0:
        return 0.0
    return producto_punto / (norma1 * norma2)

def buscar(consulta, k=5):
    """Devuelve los k documentos mas relevantes como (id, titulo, score)."""
    q = vectorizar_consulta(consulta)
    resultados = [
        (corpus[i]['id'], corpus[i]['titulo'], coseno(q, INDICE[i]))
        for i in range(len(corpus))
    ]
    resultados.sort(key=operator.itemgetter(2), reverse=True)
    return resultados[:k]
```

```python
# Prueba: deberia rankear alto d02 (sequia/maiz) y d04
for id_, titulo, score in buscar('sequia y cultivos de maiz'):
    print(f'{score:.3f}  {id_}  {titulo}')
```

## 4 · Rómpanlo

Entender dónde falla un modelo vale más que celebrarlo donde funciona. Empezamos con el caso de clase:


```python
print('Consulta: "problemas de agua"')
for id_, titulo, score in buscar('problemas de agua'):
    print(f'{score:.3f}  {id_}  {titulo}')

# Observen: d13 (agua potable) sale bien, pero d02 (crisis HIDRICA) — claramente relevante —
# queda hundido o en 0. 'agua' e 'hidrico' son cadenas distintas: TF-IDF no sabe que son sinonimos.
```

**4.a** Encuentren **2 consultas propias** donde su buscador devuelva resultados malos. Para cada una: muestren la salida y **expliquen la causa** (sinonimia, polisemia, falta de contexto, preprocesamiento agresivo, etc.). Traigan estas dos consultas a la próxima clase: con ellas abriremos BM25.

```python
print('Consulta: "temperatura alta en pacientes"')
for id_, titulo, score in buscar('temperatura alta en pacientes'):
    print(f'{score:.3f}  {id_}  {titulo}')
```

_Causa de la falla 1:_

Esta consulta busca específicamente el documento de salud (**d11**, fiebre alta por dengue). El problema es **polisemia**: el lema `alto` aparece en el corpus en dos documentos con sentidos totalmente distintos — **d02** lo usa para "zona alta" (sentido geográfico, crisis hídrica) y **d11** para "fiebre alta" (sentido médico, temperatura corporal). Como TF-IDF representa cada documento únicamente como una bolsa de tokens sin contexto, el peso que `alto` aporta a d02 es indistinguible del que aporta a d11: el modelo no sabe que son dos conceptos diferentes que casualmente comparten la misma palabra. Esto puede inflar artificialmente el score de d02 (que no tiene nada que ver con temperatura corporal) o diluir la señal real de d11, dependiendo de qué tan dominante sea ese término frente al resto de la consulta. La sinonimia (agua/hídrica) hace que cosas relacionadas no se encuentren; la polisemia hace lo inverso: que cosas no relacionadas se confundan.

```python
print('Consulta: "ola de calor por el clima"')
for id_, titulo, score in buscar('ola de calor por el clima'):
    print(f'{score:.3f}  {id_}  {titulo}')
```

_Causa de la falla 2:_

Aquí el problema es distinto: **cobertura de vocabulario**. El corpus sí tiene un documento sobre clima caliente (**d11**, "tierra caliente", dengue), pero la consulta usa palabras ("ola", "calor", "clima") que **ningún documento del corpus contiene literalmente**. No es sinonimia entre dos términos que sí están en el corpus (como agua/hídrica); es que el vocabulario del corpus simplemente nunca generó esos tokens. Un buscador TF-IDF solo puede encontrar lo que está en su índice: si una consulta usa vocabulario fuera del dominio o estilo de redacción del corpus indexado, el vector de la consulta no intersecta con ningún documento y el coseno es 0 para todos, sin importar qué tan relevante sea conceptualmente la consulta para el dominio general de noticias de Chiapas.
